```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('AAPL.csv')

# Rename column 'Close(t)' to 'Close'
df.rename(columns={'Close(t)': 'Close'}, inplace=True)

# Display the first 5 rows
print("First 5 rows:")
df.head()
```

```python
# Display the last 5 rows
print("\nLast 5 rows:")
df.tail()
```

```python
# Check the number of rows and columns
print("\nDataset shape:", df.shape)

```

```python
# Print the column names
print("\nColumn names:", df.columns.tolist())
```

# Ploting the Time Series Chart



```python
# If there is a 'Date' column, convert it to datetime and plot
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['Close'], marker='o')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('AAPL Stock Price Time Series')
    plt.grid(True)
    plt.show()
else:
    # If no Date column exists, plot using the index
    plt.figure(figsize=(10, 6))
    plt.plot(df['Close'], marker='o')
    plt.xlabel('Index')
    plt.ylabel('Close Price')
    plt.title('AAPL Stock Price Time Series')
    plt.grid(True)
    plt.show()
```

# Removing Unnecessary Columns and Create Target Variable

```python
# Remove unnecessary columns (e.g., 'Date')
if 'Date' in df.columns:
    df = df.drop(['Date'], axis=1)

# Create a new column 'Close_forecast' for the next day's close price
df['Close_forecast'] = df['Close'].shift(-1)

# Drop the last row since it will have a NaN target
df = df[:-1]

# Display the modified dataset
print("Dataset after creating 'Close_forecast' and dropping last row:")
df.head()
```

# Split the Data into Training, Validation, and Test Sets



```python
# For time series data, perform a chronological split
n = len(df)
train_size = int(n * 0.7)
val_size = int(n * 0.15)
test_size = n - train_size - val_size

print("Train size:", train_size, "Validation size:", val_size, "Test size:", test_size)

# Define the feature (using 'Close') and the target ('Close_forecast')
X = df[['Close']]
y = df['Close_forecast']

# Chronological split
X_train = X.iloc[:train_size]
y_train = y.iloc[:train_size]

X_val = X.iloc[train_size:train_size + val_size]
y_val = y.iloc[train_size:train_size + val_size]

X_test = X.iloc[train_size + val_size:]
y_test = y.iloc[train_size + val_size:]
```

# Build and Train the Linear Regression Model




```python
from sklearn.linear_model import LinearRegression

# Initialize and train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Print the coefficient and intercept of the model
print("Linear Regression Coefficient:", lr_model.coef_)
print("Linear Regression Intercept:", lr_model.intercept_)
```

# Evaluate the Model Performance



```python
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score
import math

# Function to evaluate and print performance metrics
def evaluate_model(model, X, y, dataset_name="Dataset"):
    y_pred = model.predict(X)
    mape = mean_absolute_percentage_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = math.sqrt(mse)
    r2 = r2_score(y, y_pred)
    
    print(f"Evaluation on {dataset_name}:")
    print("MAPE:", mape)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("R-Squared:", r2)
    print("-" * 40)

# Evaluate on Training Set
evaluate_model(lr_model, X_train, y_train, "Training Set")

# Evaluate on Validation Set
evaluate_model(lr_model, X_val, y_val, "Validation Set")

# Evaluate on Test Set
evaluate_model(lr_model, X_test, y_test, "Test Set")
```

# Importing Libraries and Load Data



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score, mean_absolute_error

 #Load the dataset
df_arima = pd.read_csv('AAPL.csv')

# Rename columns if needed (assuming 'Close(t)' should be 'Close')
df_arima.rename(columns={'Close(t)': 'Close'}, inplace=True)

```

```python
# Drop unnecessary columns (e.g., Date) if present
if 'Date' in df_arima.columns:
    df_arima = df_arima.drop(['Date'], axis=1)

# Define target variable (Close)
target = df_arima['Close']

print("Dataset head:")
df_arima.head()
```

# Chronological Data Split (88% Train, 10% Validation, 2% Test)



```python
# Total number of observations
n = len(df_arima)
train_size = int(0.88 * n)
val_size = int(0.10 * n)
test_size = n - train_size - val_size

print("Total observations:", n)
print("Training size:", train_size)
print("Validation size:", val_size)
print("Test size:", test_size)

# Split the target series
train_data = target.iloc[:train_size]
val_data = target.iloc[train_size:train_size + val_size]
test_data = target.iloc[train_size + val_size:]
```

# Defining and Fit the ARIMA Model



```python
# Set initial ARIMA parameters (p, d, q)
p, d, q = 1, 1, 1

# Define and fit the ARIMA model on training data
model_arima = ARIMA(train_data, order=(p, d, q))
model_fit = model_arima.fit()

# Print ARIMA model summary
print(model_fit.summary())
```

# Generate Predictions and Evaluate Model Performance


```python
# In-sample prediction for training data
pred_train = model_fit.predict(start=0, end=train_size - 1)

# Forecast for validation and test sets (total steps = validation + test)
forecast_steps = val_size + test_size
forecast = model_fit.forecast(steps=forecast_steps)

# Split the forecast into validation and test predictions
pred_val = forecast[:val_size]
pred_test = forecast[val_size:]

# Define a function to evaluate predictions
def evaluate_forecast(actual, predicted, dataset_name="Dataset"):
    mape = mean_absolute_percentage_error(actual, predicted)
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    r2 = r2_score(actual, predicted)
    mae = mean_absolute_error(actual, predicted)
    print(f"Evaluation Metrics for {dataset_name}:")
    print(f"MAPE: {mape:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"R-Squared: {r2:.4f}")
    print("-" * 40)

# Evaluate on Training data
evaluate_forecast(train_data, pred_train, "Training Data")

# Evaluate on Validation data
evaluate_forecast(val_data, pred_val, "Validation Data")

# Evaluate on Test data
evaluate_forecast(test_data, pred_test, "Test Data")
```

# Plot Actual vs. Predicted Values



```python
# Plot Actual vs Predicted for the Validation Set
plt.figure(figsize=(10, 5))
plt.plot(val_data.index, val_data, label='Actual Validation')
plt.plot(val_data.index, pred_val, label='Predicted Validation', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.title('ARIMA Model: Actual vs Predicted (Validation Set)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Actual vs Predicted for the Test Set
plt.figure(figsize=(10, 5))
plt.plot(test_data.index, test_data, label='Actual Test')
plt.plot(test_data.index, pred_test, label='Predicted Test', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.title('ARIMA Model: Actual vs Predicted (Test Set)')
plt.legend()
plt.grid(True)
plt.show()
```

# Analyzing Model Residuals



```python
# Calculate residuals on training data
residuals = train_data - pred_train

# Plot residuals over time
plt.figure(figsize=(10, 5))
plt.plot(residuals.index, residuals)
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residuals of ARIMA Model on Training Data')
plt.grid(True)
plt.show()

# Plot a histogram of residuals
plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=20)
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True)
plt.show()
```

# Adjust ARIMA Parameters and Refit (Optional)



```python
# Example: Adjust ARIMA parameters to (2,1,2) and refit the model
p_new, d_new, q_new = 2, 1, 2
model_arima_new = ARIMA(train_data, order=(p_new, d_new, q_new))
model_fit_new = model_arima_new.fit()

print("ARIMA Model Summary with parameters (2,1,2):")
print(model_fit_new.summary())

# Generate new forecasts for validation and test sets using the updated model
forecast_new = model_fit_new.forecast(steps=forecast_steps)
pred_val_new = forecast_new[:val_size]
pred_test_new = forecast_new[val_size:]

# Evaluate the new model performance on validation and test data
evaluate_forecast(val_data, pred_val_new, "Validation Data (New ARIMA)")
evaluate_forecast(test_data, pred_test_new, "Test Data (New ARIMA)")
```

# (Optional) Grid Search to Optimize ARIMA Parameters

```python
# Optional: A simple grid search to find the best ARIMA parameters based on validation MSE.
import warnings
warnings.filterwarnings("ignore")

p_values = [0, 1, 2]
d_values = [0, 1]
q_values = [0, 1, 2]

best_score, best_cfg = float("inf"), None

for p in p_values:
    for d in d_values:
        for q in q_values:
            order = (p, d, q)
            try:
                model = ARIMA(train_data, order=order)
                model_fit = model.fit()
                # Forecast for the validation period
                forecast_temp = model_fit.forecast(steps=val_size)
                mse = mean_squared_error(val_data, forecast_temp)
                print(f"ARIMA{order} MSE = {mse:.4f}")
                if mse < best_score:
                    best_score, best_cfg = mse, order
            except Exception as e:
                continue

print(f"\nBest ARIMA parameters: {best_cfg} with Validation MSE = {best_score:.4f}")
```

# Final Evaluation on Test Data Using the Optimized Model


```python
# If grid search provided the best configuration, refit the model with those parameters
if best_cfg is not None:
    model_final = ARIMA(train_data, order=best_cfg)
    model_fit_final = model_final.fit()
    forecast_final = model_fit_final.forecast(steps=forecast_steps)
    # Split forecasts: first part for validation, second part for test
    pred_test_final = forecast_final[val_size:]
    
    print("Final Model Performance on Test Data:")
    evaluate_forecast(test_data, pred_test_final, "Final Test Data")
```
